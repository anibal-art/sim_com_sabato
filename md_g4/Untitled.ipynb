```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("energia.dat", 
                 delim_whitespace=True, 
                 header=None, 
                 names=["step", "energy"])
print(len(df))
plt.plot(df["step"], df["energy"], ls='',marker='.')
plt.xlabel("step")
plt.ylabel("energy")
# plt.xlim(10,10000)
# plt.ylim(-1000,10)
# plt.yscale("log")
plt.xscale("log")
plt.tight_layout()
plt.show()
```
