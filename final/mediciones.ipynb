```python
import MDAnalysis as mda


u = mda.Universe(
    "/home/anibal/sim_com_sabato/final/dump.4000atoms_T2p5.lammpstrj",
    format="LAMMPSDUMP",
)

from MDAnalysis.analysis import rdf

all_atoms = u.select_atoms("all")

gr = rdf.InterRDF(
    all_atoms, all_atoms,
    nbins=100,
    range=(0.0, 8.0),
)
gr.run(start=20, step=1)
```

```python

import matplotlib.pyplot as plt
%matplotlib widget
plt.plot(gr.bins, gr.rdf)
plt.ylim(-1,200)
plt.title("g(r) del fluido a T=1")
plt.xlabel(r"r (unidades reducidas)")
plt.ylabel(r"g(r)")
plt.tight_layout()
plt.savefig("/home/anibal/sim_com_sabato/final_presentation/figures_lecture/gr_T1_gota.png")
plt.show()
```
